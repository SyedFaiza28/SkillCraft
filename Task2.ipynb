```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the CSV file (update the filename if different)
df = pd.read_csv("C:\\Users\\DELL\\Desktop\\country_wise_latest.csv")
```

```python
# Print column names to verify
print("Columns in CSV:", df.columns)
```

```python
# Clean column names (remove leading/trailing spaces)
df.columns = df.columns.str.strip()
```

```python
# Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
# Exploratory Data Analysis (EDA)
# Summary Statistics
print(df.describe(include='all'))
```

```python
# Check if 'Country/Region' and 'Confirmed' exist
if 'Country/Region' in df.columns and 'Confirmed' in df.columns:
    
    # Sort by Confirmed cases for better visualization
    df_sorted = df.sort_values(by='Confirmed', ascending=False).head(10)

    # Plot top 10 countries by confirmed cases
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_sorted, x='Country/Region', y='Confirmed', palette="viridis")
    plt.title('Top 10 Countries by Confirmed COVID-19 Cases')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Required columns not found. Please check the CSV structure")
```

```python
# Top 10 Countries by Total Deaths
if 'Country/Region' in df.columns and 'Deaths' in df.columns:
    df_deaths = df.sort_values(by='Deaths', ascending=False).head(10)
    
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_deaths, x='Country/Region', y='Deaths', palette='Reds')
    plt.title('Top 10 Countries by COVID-19 Deaths')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Column 'Deaths' not found.")
```

```python
# Top 10 Countries by Total Recovered
if 'Country/Region' in df.columns and 'Recovered' in df.columns:
    df_recovered = df.sort_values(by='Recovered', ascending=False).head(10)
    
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_recovered, x='Country/Region', y='Recovered', palette='Greens')
    plt.title('Top 10 Countries by COVID-19 Recoveries')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Column 'Recovered' not found.")
```
