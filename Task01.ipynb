```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
data = {
    'Age': [22, 25, 47, 52, 46, 35, 23, 31, 44, 28, 38, 40, 29, 30, 34, 36, 41, 45, 50, 60],
    'Salary': [30000, 32000, 47000, 55000, 52000, 40000, 31000, 36000, 49000, 33000,
               42000, 45000, 34000, 37000, 39000, 41000, 43000, 46000, 53000, 60000],
    'Experience': [1, 2, 10, 12, 11, 7, 2, 4, 9, 3, 6, 7, 3, 4, 5, 6, 8, 9, 11, 15]
}

df = pd.DataFrame(data)
```

```python
# Plotting histograms for each numerical column
plt.figure(figsize=(14, 8))  
```

```python
# Histogram for Age
plt.subplot(3, 1, 1)
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title('Age Distribution')

plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
# Histogram for Salary
plt.subplot(3, 1, 2)
sns.histplot(df['Salary'], bins=10, kde=True, color='lightgreen')
plt.title('Salary Distribution')

plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
# Histogram for Experience
plt.subplot(3, 1, 3)
sns.histplot(df['Experience'], bins=10, kde=True, color='salmon')
plt.title('Experience Distribution')

plt.tight_layout()  # Adjust spacing
plt.show()
```
